```python
from langchain_community.tools.tavily_search import TavilySearchResults

search = TavilySearchResults()

results = search.invoke("Detailed report on Delhi attack.")

print(results)
```

```python
news_title = [items['title'] for items in results]
print(news_title[0])
print(news_title[1])
print(news_title[2])
```

```python
print(news_title[0])
print("\n")

main_content = [items['content'] for items in results]

print(main_content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# Replace with your LM Studio server's base URL and port
lm_studio_base_url = "http://127.0.0.1:1234" 

# Initialize the ChatOpenAI model
# You might need to specify the model name if LM Studio requires it, 
# but often it defaults to the loaded model.
llm = ChatOpenAI(
    openai_api_base=lm_studio_base_url,
    model_name="deepseek/deepseek-r1-0528-qwen3-8b", # Replace with the actual model name if needed, or a placeholder
    temperature=0.7 # Adjust as desired
)

# Example usage:
messages = [
    HumanMessage(
        content="Explain the concept of quantum entanglement in simple terms."
    )
]
response = llm.invoke(messages)
print(response.content)
```
