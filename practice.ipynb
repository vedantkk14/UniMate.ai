```python
def chat_model():
    llm = HuggingFaceEndpoint(
        repo_id='litert-community/Gemma3-1B-IT',
        task='text-generation',
        temperature=0.4,
        max_new_tokens=100
    )

    return ChatHuggingFace(llm = llm)
```
